```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,pow):
    return x**pow
```

```python
plt.figure(figsize=(5,5))
x = np.linspace(0.001,1,1000)
for p in range(1,11):
    if p==0:
        continue
    y=f(x,p)
    plt.scatter(x,y)
    y=f(x,1/p)
    plt.scatter(x,y)
plt.show()
```
