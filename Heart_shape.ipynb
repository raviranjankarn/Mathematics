```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
<img src="Heart_curve_function.jpg">
```

```python
from math import sqrt
```

```python
def f(x):
    return sqrt(1-x**2)+x**(2/3)
def g(x):
    return -sqrt(1-x**2)+x**(2/3)
```

```python
x = np.linspace(0,1,1000)
vf = np.vectorize(f)
wf = np.vectorize(g)
y1=vf(x)
y2=wf(x )

plt.figure(figsize=(5,5))
plt.scatter(x,y1)
plt.scatter(x,y2)
plt.scatter(-x,y1)
plt.scatter(-x,y2)
plt.show()
```
