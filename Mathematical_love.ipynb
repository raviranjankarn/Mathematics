```python
import numpy as np
import math
import matplotlib.pyplot as plt
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
x1= np.linspace(0.01,5,1000)
x2=np.linspace(-5,5,1000)
x3=np.linspace(-1*np.pi,np.pi,1000)
y1=1/x1
y2=np.sqrt(9-x2**2)
y3=-np.sqrt(9-x2**2)
y4=abs(-2*x2)
y5= [-3*abs(math.sin(x)) for x in x3]
```

```python
fig, ax = plt.subplots(1,4)
plt.setp(ax, xticks=[], yticks=[])
ax[0].plot(x1,y1,'r')
ax[1].plot(x2,y2,'r')
ax[1].plot(x2,y3,'r')
ax[2].plot(x2,y4,'r')
ax[3].plot(y5,x2,'r')
plt.show()
```
